Criando o banco SQLite

```python
import sqlite3


# Connect to the database
conn = sqlite3.connect('banco_de_dados.db')

# Create Pessoa table
conn.execute('''CREATE TABLE Pessoa
                (CPF TEXT  NOT NULL,
                primeiro_nome TEXT NOT NULL,
                nome_do_meio TEXT,
                sobrenome TEXT NOT NULL,
                idade INTEGER NOT NULL,
                conta TEXT NOT NULL,
                FOREIGN KEY (conta) REFERENCES Conta (numero))''')

# Create Conta table
conn.execute('''CREATE TABLE Conta
                (Agência TEXT NOT NULL,
                Número TEXT  NOT NULL,
                Saldo REAL NOT NULL,
                Gerente TEXT NOT NULL,
                Titular TEXT NOT NULL)''')
# Commit changes and close connection
conn.commit()
conn.close() 
```

O script envia dados do arquivo txt para a tabela pessoa do banco de dados sqlite

```python
import sqlite3

# Open the file in read mode
with open("nome.txt", "r") as file:
    data = file.read()
    # Count the number of commas in the file
    num_commas = data.count(",")
    x = 0
    # Loop through the number of commas found
    for i in range(num_commas):
        x += 1
        if x == 6:
            with open("nome.txt", "r") as file_in, open("New Text Document (2).txt", "w") as file_out:
                # Lê todas as linhas do arquivo de entrada
                for line in file_in:
                    # Divide a linha em partes separadas por vírgula
                    parts = line.split(",")
                    # Recorta as cinco primeiras partes
                    new_line = ",".join(parts[:6])
                    # Escreve a nova linha no arquivo de saída
                    file_out.write(new_line + "\n")
                    # Reabre o arquivo de entrada no modo de gravação
                    with open("nome.txt", "w") as file_in2:
                        # Recorta a linha original
                        line = ",".join(parts[6:])
                        # Reescreve a linha original no arquivo de entrada
                        file_in2.write(line)
                    # Divide a nova linha em campos separados por vírgula
                    fields = new_line.strip().split(',')
                    # Verifica se há seis campos
                    if len(fields) == 6:
                        conn = sqlite3.connect('banco_de_dados.db')
                        # Descompacta os valores
                        CPF, primeiro_nome, nome_do_meio, sobrenome, idade, conta = fields
                        print(fields)
                        # Insere os valores no banco de dados
                        conn.execute("INSERT INTO Pessoa (CPF, primeiro_nome, nome_do_meio, sobrenome, idade, conta) VALUES (?, ?, ?, ?, ?, ?)", (CPF, primeiro_nome, nome_do_meio, sobrenome, idade, conta))
                        conn.commit()
                        conn.close()
                file_in.close() 
                file_out.close()
                file_in2.close()              
            x -=6 
    file.close()
with open('nome.txt', 'r') as f:
    conn = sqlite3.connect('banco_de_dados.db')
    for line in f:
        CPF, primeiro_nome, nome_do_meio, sobrenome, idade, conta = line.strip().split(',')
        print(line)
        conn.execute("INSERT INTO Pessoa (CPF, primeiro_nome, nome_do_meio, sobrenome, idade, conta) VALUES (?, ?, ?, ?, ?, ?)", (CPF, primeiro_nome, nome_do_meio, sobrenome, idade, conta))
        conn.commit()   
    conn.close() 
    f.close()    
```

Script python que   adiciona, remove e edita qualquer linha no banco de dados SQLite via input. Realiza consultas no banco de dados SQLite e salva o resultado em novos arquivos de texto, organiza as consultas por tipo em diferentes pastas (ex. consulta por nome, por saldo, por idade, etc.), se a pasta não existir, crie uma nova, ler e imprimir resultados de consulta através de arquivos criados.

```python
import sqlite3
import os
con = sqlite3.connect('banco_de_dados.db')
class dados():
    def add_pessoa():
        CPF = input('Digite o CPF da pessoa: ')
        primeiro_nome = input('Digite o nome da pessoa: ')
        nome_do_meio = input('Digite o nome do meio da pessoa: ')
        sobrenome = input('Digite o sobrenome da pessoa: ')
        idade = int(input('Digite a idade da pessoa: '))
        conta = input('Digite a conta da pessoa: ')
        con.execute("INSERT INTO Pessoa (CPF, primeiro_nome, nome_do_meio, sobrenome, idade, conta) VALUES (?, ?, ?, ?, ?, ?)", (CPF, primeiro_nome, nome_do_meio, sobrenome, idade, conta))
        con.commit()
        print('Pessoa adicionada com sucesso.')

    def add_conta():
        Agência = input('Digite a agência da conta: ')
        Número = input('Digite o número da conta: ')
        Saldo = float(input('Digite o saldo da conta: '))
        Gerente = input('Digite o nome do gerente da conta: ')
        Titular = input('Digite o CPF do titular da conta: ')
        con.execute("INSERT INTO Conta (Agência, Número, Saldo, Gerente, Titular) VALUES (?, ?, ?, ?, ?)", (Agência, Número, Saldo, Gerente, Titular))
        con.commit()
        print('Conta adicionada com sucesso.')
    def remove_pessoa():
        CPF = input('Digite o CPF da pessoa que deseja remover: ')
        con.execute("DELETE FROM Pessoa WHERE CPF=?", (CPF,))
        con.commit()
        print('Pessoa removida com sucesso.')
    def remove_conta():
        Número = input('Digite o número da conta que deseja remover: ')
        con.execute("DELETE FROM Conta WHERE Número=?", (Número,))
        con.commit()
        print('Conta removida com sucesso.')

    def edit_pessoa():
        CPF = input('Digite o CPF da pessoa que deseja editar: ')
        primeiro_nome = input('Digite o novo nome da pessoa: ')
        nome_do_meio = input('Digite o novo nome do meio da pessoa: ')
        sobrenome = input('Digite o novo sobrenome da pessoa: ')
        idade = int(input('Digite a nova idade da pessoa: '))
        novo_CPF = input('Digite o novo CPF: ')
        conta = input('Digite a nova conta da pessoa: ')
        con.execute("UPDATE Pessoa SET primeiro_nome=?, nome_do_meio=?, CPF=?, sobrenome=?, idade=?, conta=? WHERE CPF=?", (primeiro_nome, nome_do_meio, novo_CPF, sobrenome, idade, conta, CPF))
        con.commit()
        print('Pessoa atualizada com sucesso.')

    def edit_conta():
        Número = input('Digite o número da conta da pessoa que deseja editar: ')
        Agência = input('Digite o nova agência da pessoa: ')
        Novo_Número = input('Digite a novo número da conta da pessoa: ')
        Titular = input('Digite o titular da conta: ')
        Saldo = float(input('Digite o saldo da conta: '))
        Gerente = input('Digite o novo gerente da pessoa: ')
        con.execute("UPDATE Conta SET Agência=?, Número=?, Saldo=?, Gerente=?, Titular=? WHERE Número=?", (Agência, Novo_Número, Saldo, Gerente, Titular, Número))
        con.commit()
        print('Conta atualizada com sucesso.')   
    def consulta_nome():
        primeiro_nome = input('Digite o nome a ser consultado: ')
        cur = con.cursor()
        cur.execute("SELECT * FROM Pessoa WHERE primeiro_nome=?", (primeiro_nome,))
        rows = cur.fetchall()
        if len(rows) == 0:
            print('Não há registros com esse primeiro nome.')
        else:
            pasta = 'consultas/primeiro_nome/'
            if not os.path.exists(pasta):
                os.makedirs(pasta)
            arquivo = pasta + str(primeiro_nome) + '.txt'
            with open(arquivo, 'w') as f:
                for row in rows:
                    f.write(f"CPF: {row[0]}\n")
                    f.write(f"Nome: {row[1]} {row[2]} {row[3]}\n")
                    f.write(f"Idade: {row[4]}\n")
                    f.write(f"Conta: {row[5]}\n\n")
            print(f'Resultados salvos em {arquivo}.')
    def consulta_saldo():
        Saldo = float(input('Digite o saldo a ser consultado: '))
        cur = con.cursor()
        cur.execute("SELECT * FROM Conta WHERE Saldo=?", (Saldo,))
        rows = cur.fetchall()
        if len(rows) == 0:
            print('Não há registros com esse saldo.')
        else:
            pasta = 'consultas/saldo/'
            if not os.path.exists(pasta):
                os.makedirs(pasta)
            arquivo = pasta + str(Saldo) + '.txt'
            with open(arquivo, 'w') as f:
                for row in rows:
                    f.write(f"Agência: {row[0]}\n")
                    f.write(f"Número: {row[1]}\n")
                    f.write(f"Saldo: {row[2]}\n")
                    f.write(f"Gerente: {row[3]}\n")
                    f.write(f"Titular: {row[4]}\n\n")
            print(f'Resultados salvos em {arquivo}.')
    def consulta_idade():
        idade = input('Digite a idade a ser consultado: ')
        cur = con.cursor()
        cur.execute("SELECT * FROM Pessoa WHERE idade=?", (idade,))
        rows = cur.fetchall()
        if len(rows) == 0:
            print('Não há registros com esse idade.')
        else:
            pasta = 'consultas/idade/'
            if not os.path.exists(pasta):
                os.makedirs(pasta)
            arquivo = pasta + idade + '.txt'
            with open(arquivo, 'w') as f:
                for row in rows:
                    f.write(f"CPF: {row[0]}\n")
                    f.write(f"Nome: {row[1]} {row[2]} {row[3]}\n")
                    f.write(f"Idade: {row[4]}\n")
                    f.write(f"Conta: {row[5]}\n\n")
            print(f'Resultados salvos em {arquivo}.')
    def consulta_Gerente():
        Gerente = input('Digite o saldo a ser consultado: ')
        cur = con.cursor()
        cur.execute("SELECT * FROM Conta WHERE Gerente=?", (Gerente,))
        rows = cur.fetchall()
        if len(rows) == 0:
            print('Não há registros com esse Gerente.')
        else:
            pasta = 'consultas/gerente/'
            if not os.path.exists(pasta):
                os.makedirs(pasta)
            arquivo = pasta + str(Gerente) + '.txt'
            with open(arquivo, 'w') as f:
                for row in rows:
                    f.write(f"Agência: {row[0]}\n")
                    f.write(f"Número: {row[1]}\n")
                    f.write(f"Saldo: {row[2]}\n")
                    f.write(f"Gerente: {row[3]}\n")
                    f.write(f"Titular: {row[4]}\n\n")
            print(f'Resultados salvos em {arquivo}.')
    def ler_imprimir_gerente():
        resultado = input("Digite o nome do arquivo destro da pasta gerente: ")
        with open('consultas/gerente/' + resultado + '.txt', 'r') as arquivo:
            conteudo = arquivo.read()
            print(conteudo)
    def ler_imprimir_idade():
        resultado = input("Digite o nome do arquivo destro da pasta idade: ")
        with open('consultas/idade/' + resultado + '.txt', 'r') as arquivo:
            conteudo = arquivo.read()
            print(conteudo)
    def ler_imprimir_nome():
        resultado = input("Digite o nome do arquivo destro da pasta nome: ")
        with open('consultas/nome/' + resultado + '.txt', 'r') as arquivo:
            conteudo = arquivo.read()
            print(conteudo)
    def ler_imprimir_saldo():
        resultado = input("Digite o nome do arquivo destro da pasta saldo:")
        with open('consultas/saldo/' + resultado + '.txt', 'r') as arquivo:
            conteudo = arquivo.read()
            print(conteudo)                        
    
while True:
    print("Selecione uma opção:")
    print("1 - Adicionar dados na tabela Pessoa:")
    print("2 - Adicionar dados na tabela Conta:")
    print("3 - Remover dados na tabela Pessoa:")
    print("4 - Remover dados na tabela Conta:")
    print("5 - Editar dados na tabela Pessoa:")
    print("6 - Editar dados na tabela Conta")
    print("7 - Consultar por Nome")
    print("8 - Consultar por Saldo")
    print("9 - Consultar por Idade")
    print("10 - Consultar por Gerente")
    print("11 - ler e imprimir arquivos txt da pasta gerente ")
    print("12 - ler e imprimir arquivos txt da pasta idade ")
    print("13 - ler e imprimir arquivos txt da pasta nome ")
    print("14 - ler e imprimir arquivos txt da pasta saldo ")
    print("0 - Sair do programa")

    opcao = input()

    if opcao == "1":
        dados.add_pessoa()



    elif opcao == "2": 
        dados.add_conta()


    elif opcao == "3":
        dados.remove_pessoa()


    elif opcao == "4":
        dados.remove_conta()

    elif opcao == "5":
        dados.edit_pessoa()

    elif opcao == "6":
        dados.edit_conta()


    elif opcao == "7":
        dados.consulta_nome()


    elif opcao == "8":
        dados.consulta_saldo()

    elif opcao == "9":
        dados.consulta_idade()

    elif opcao == "10":
        dados.consulta_Gerente()

    elif opcao == "11":
        dados.ler_imprimir_gerente()

    elif opcao == "12":
        dados.ler_imprimir_idade()

    elif opcao == "13":
        dados.ler_imprimir_nome()

    elif opcao == "14":
        dados.ler_imprimir_saldo()

    elif opcao == "0":
        print("Obrigado por utilizar a biblioteca. Até mais!")
        break

    else:
        print("Opção inválida.")    
```
